# Merging

```python
import os
import xarray as xr
from tqdm import tqdm
import dask

# Directory where your .nc files are located (replace with your actual directory path)
data_dir = r"C:\Users\Majid\Desktop\Extreem Projections\Data Processing\Loading and Merging Datasets\10years-Test\test1"

# Get the last 5 sections of the directory path
sections = data_dir.split(os.path.sep)[-5:]
last_sections = "_".join(sections)

# Replace any non-alphanumeric characters with underscores
file_name = "".join(c if c.isalnum() else "_" for c in last_sections)

# Get a list of all .nc files in the directory
files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith('.nc')]

# Create an empty list to hold all datasets
datasets = []

## memory error
# Open datasets lazily with dask
for file in tqdm(files, desc='Loading files', unit='file'):
    datasets.append(xr.open_dataset(file, chunks={'time': 'auto'}))

# Concatenate datasets along the time dimension using dask
combined_dataset = xr.concat(datasets, dim='time')

# Save the combined dataset to a new NetCDF file with dynamically generated name
combined_dataset.to_netcdf(os.path.join(data_dir, f'combined_{file_name}.nc'))

# Close all individual datasets
for ds in datasets:
    ds.close()

# Close the combined dataset
combined_dataset.close()

print(f"Combined dataset saved successfully as 'combined_{file_name}.nc'.")
```

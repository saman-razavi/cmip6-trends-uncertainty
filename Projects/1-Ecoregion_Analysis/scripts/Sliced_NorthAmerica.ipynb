```python
import os
import glob
import xarray as xr

# Function to slice and save NetCDF file
def slice_and_save_nc(input_nc_file, lat_range, lon_range, output_folder, total_files):
    # Load the dataset
    ds = xr.open_dataset(input_nc_file)
    
    # Slice the dataset
    ds_subset = ds.sel(lat=lat_range, lon=lon_range)
    
    # Generate output file name
    file_name = os.path.basename(input_nc_file)
    
    # Get relative path of input file
    relative_path = os.path.relpath(input_nc_file, input_folder)
    
    # Create corresponding subfolder in output directory
    output_subfolder = os.path.join(output_folder, os.path.dirname(relative_path))
    os.makedirs(output_subfolder, exist_ok=True)
    
    # Construct output file path within the subfolder
    output_nc_file = os.path.join(output_subfolder, f"{file_name.replace('.nc', '_NorthAmerica.nc')}")
    
    # Save the subsetted dataset to a new NetCDF file
    ds_subset.to_netcdf(output_nc_file)
    
    # Optionally, close the dataset
    ds.close()
    
    # Print progress
    current_file_index = slice_and_save_nc.counter + 1
    progress = (current_file_index / total_files) * 100
    print(f"Processed file {current_file_index}/{total_files} ({progress:.2f}%): {file_name}")
    
    # Update counter
    slice_and_save_nc.counter += 1

# Counter initialization for function attribute
slice_and_save_nc.counter = 0

# Path to the input folder containing NetCDF files
input_folder = r'C:\Users\Majid\Desktop\Extreme Projections\DATA sets\10years-Test\CanESM'

# Path to the output folder where sliced files will be saved
output_folder = r'C:\Users\Majid\Desktop\Extreme Projections\DATA sets\10years-Test\CanESM-sliced'

# Define the latitude and longitude ranges for slicing
lat_range = slice(14, 84)
lon_range = slice(180, 308)

# Find all NetCDF files in the input folder and its subfolders
nc_files = glob.glob(os.path.join(input_folder, '**', '*.nc'), recursive=True)
total_files = len(nc_files)

# Process each NetCDF file
for nc_file in nc_files:
    slice_and_save_nc(nc_file, lat_range, lon_range, output_folder, total_files)
```
